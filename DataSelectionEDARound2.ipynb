##Implementing Hyperparameter tuning

```python
##ORIGINAL modified

from sklearn.model_selection import ParameterGrid
params_grid = {
               'seasonality_prior_scale': [.01, .05, .1, .25, 1, 5, 10],
              'changepoint_prior_scale':[0.001, 0.01, 0.1, 0.25, 0.3, 0.4, 0.5]
               }

grid = ParameterGrid(params_grid)
cnt = 0
for p in grid:
    cnt = cnt+1

print('Total Possible Models',cnt)
```

```python
strt='2021-03-01'
end='2021-09-01'
model_parameters = pd.DataFrame(columns = ['MAPE','Parameters'])
for p in grid:
    test = pd.DataFrame()
    print(p)
    # random.seed(0)

    #The only parameters that we are tuning is changepoint_prior_scale, seasonality_prior_scale, and seasonality mode we are setting to multiplicative

    train_model =Prophet(changepoint_prior_scale = p['changepoint_prior_scale'],
                     
                         seasonality_mode = 'multiplicative',
                         seasonality_prior_scale = p['seasonality_prior_scale'],
                         n_changepoints= 1,
                         weekly_seasonality= False,
                         daily_seasonality = False,
                         interval_width=0.85)
    
    train_model = train_model.add_seasonality(
         name = 'weekly',
         period=7,
         fourier_order = 20
    )


    train_model = train_model.add_seasonality(
        name = 'daily',
        period=1,
        fourier_order = 15
        )
    

    # train_model.add_country_holidays(country_name='US')
    train_model.fit(train_df_champaign)
    train_forecast = train_model.make_future_dataframe(periods= 6, freq='M', include_history = False)
    train_forecast = train_model.predict(train_forecast)
    test = train_forecast[['ds','yhat']]
    Actual = agg_rev_by_month_champaign[(agg_rev_by_month_champaign['ds']>strt) & (agg_rev_by_month_champaign['ds']<=end)]
    MAPE = mean_absolute_percentage_error(Actual['y'],abs(test['yhat']))
    print('Mean Absolute Percentage Error(MAPE)------------------------------------',MAPE)

    model_parameters = model_parameters.append({'MAPE':MAPE,'Parameters':p},ignore_index = True)
```

```python
parameters_champaign_1 = model_parameters.sort_values(by=['MAPE']).reset_index(drop = True)
# parameters_comm_1 = parameters_comm_1.reset_index(drop=True)
parameters_champaign_1.head()
```

```python
MAPE_champaign_hyptertest1 = parameters_champaign_1['MAPE'][0]
parameters_champaign_test1 = parameters_champaign_1['Parameters'][0]
print(MAPE_champaign_hyptertest1)
print(parameters_champaign_test1)
```

##Implementing Hyperparameter tuning

```python
from sklearn.model_selection import ParameterGrid
params_grid = {
               'seasonality_prior_scale': [.01, .05, .1, .25, 1, 5, 10],
               'changepoint_prior_scale':[0.001, 0.01, 0.1, 0.25, 0.3, 0.4, 0.5]
               }

grid = ParameterGrid(params_grid)
cnt = 0
for p in grid:
    cnt = cnt+1

print('Total Possible Models',cnt)
```

```python
strt='2021-03-01'
end='2021-09-01'
model_parameters = pd.DataFrame(columns = ['MAPE','Parameters'])
for p in grid:
    test = pd.DataFrame()
    print(p)
    # random.seed(0)

    #The only parameters that we are tuning is changepoint_prior_scale, seasonality_prior_scale, and seasonality mode we are setting to multiplicative

    train_model =Prophet(changepoint_prior_scale = p['changepoint_prior_scale'],
                         seasonality_mode = 'multiplicative',
                        #  seasonality_prior_scale = p['seasonality_prior_scale'],
                         n_changepoints= 1,
                         weekly_seasonality= False,
                         daily_seasonality = False,
                         interval_width=0.85)
    
    train_model = train_model.add_seasonality(
         name = 'weekly',
         period=7,
         fourier_order = 20
    )


    train_model = train_model.add_seasonality(
        name = 'daily',
        period=1,
        fourier_order = 15
        )
    

    # train_model.add_country_holidays(country_name='US')
    train_model.fit(train_df_effingham)
    train_forecast = train_model.make_future_dataframe(periods= 6, freq='M', include_history = False)
    train_forecast = train_model.predict(train_forecast)
    test = train_forecast[['ds','yhat']]
    Actual = agg_rev_by_month_effingham[(agg_rev_by_month_effingham['ds']>strt) & (agg_rev_by_month_effingham['ds']<=end)]
    MAPE = mean_absolute_percentage_error(Actual['y'],abs(test['yhat']))
    print('Mean Absolute Percentage Error(MAPE)------------------------------------',MAPE)

    model_parameters = model_parameters.append({'MAPE':MAPE,'Parameters':p},ignore_index=True)
```

```python
parameters_effingham_1 = model_parameters.sort_values(by=['MAPE']).reset_index(drop = True)
# parameters_comm_1 = parameters_comm_1.reset_index(drop=True)
parameters_effingham_1.head()
```

##Implementing Hyperparameter tuning

```python
#{'changepoint_prior_scale': 0.5, 'seasonality_prior_scale': 0.1} -> this gave us the best score originially


from sklearn.model_selection import ParameterGrid
params_grid = {
               'seasonality_prior_scale': [.01, .05, .1, .25, 1, 5, 10],
              'changepoint_prior_scale':[0.001, 0.01, 0.1, 0.25, 0.3, 0.4, 0.5]
               }

grid = ParameterGrid(params_grid)
cnt = 0
for p in grid:
    cnt = cnt+1

print('Total Possible Models',cnt)
```

Implementing Hyperparameter tuning Test 1

```python

strt='2021-03-01'
end='2021-09-01'
model_parameters = pd.DataFrame(columns = ['MAPE','Parameters'])
for p in grid:
    test = pd.DataFrame()
    print(p)
    # random.seed(0)

    #The only parameters that we are tuning is changepoint_prior_scale, seasonality_prior_scale, and seasonality mode we are setting to multiplicative

    train_model =Prophet(changepoint_prior_scale = p['changepoint_prior_scale'],
                     
                         seasonality_mode = 'multiplicative',
                         seasonality_prior_scale = p['seasonality_prior_scale'],
                         n_changepoints= 1,
                         weekly_seasonality= False,
                         daily_seasonality = False,
                         interval_width=0.85)
    
    train_model = train_model.add_seasonality(
         name = 'weekly',
         period=7,
         fourier_order = 20
    )


    train_model = train_model.add_seasonality(
        name = 'daily',
        period=1,
        fourier_order = 15
        )
    

    # train_model.add_country_holidays(country_name='US')
    train_model.fit(train_df_farmers)
    train_forecast = train_model.make_future_dataframe(periods= 6, freq='M', include_history = False)
    train_forecast = train_model.predict(train_forecast)
    test = train_forecast[['ds','yhat']]
    Actual = agg_rev_by_month_farmers[(agg_rev_by_month_farmers['ds']>strt) & (agg_rev_by_month_farmers['ds']<=end)]
    MAPE = mean_absolute_percentage_error(Actual['y'],abs(test['yhat']))
    print('Mean Absolute Percentage Error(MAPE)------------------------------------',MAPE)

    model_parameters = model_parameters.append({'MAPE':MAPE,'Parameters':p},ignore_index=True)
```

```python
parameters_farmers_1 = model_parameters.sort_values(by=['MAPE']).reset_index(drop = True)
# parameters_comm_1 = parameters_comm_1.reset_index(drop=True)
parameters_farmers_1.head()
```

```python
MAPE_farmers_hyptertest1 = parameters_farmers_1['MAPE'][0]
parameters_farmers_test1 = parameters_farmers_1['Parameters'][0]
print(MAPE_farmers_hyptertest1)
print(parameters_farmers_test1)
```

```python
print('Best combination of parameters after performing Hyperparameter optimization Test 1 for Farmers Branch, TX ', parameters_farmers_test1)
print('MAPE score of best combination of parameters For Farmers Branch, TX', round(MAPE_farmers_hyptertest1, 5))
```
